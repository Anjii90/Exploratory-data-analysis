```python
df = df.dropna(subset=['name', 'host_name'])
print(df.isnull().sum())
```

```python
print(df.isnull().sum())
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
df['last_review_year'] = df['last_review'].dt.year
```

**Neighbourhood Group Popularity**

```python
neighbourhood_counts = df['neighbourhood_group'].value_counts()
print(neighbourhood_counts)
```

Insight: Manhattan has the highest number of listings (21,643), followed by Brooklyn (20,084). These areas are the most popular among hosts.

**Room Type Preferences**

```python
room_type_counts = df['room_type'].value_counts()
print(room_type_counts)
```

Insight: Entire homes/apartments make up the majority of listings (25,393), indicating a strong preference for private accommodations.

**Price Distribution**

```python
price_stats = df['price'].describe()
print(price_stats)
```

The median listing price is $106     per     night,     with     the majority   of listings priced  between $69 and $175.

# **Correaltions**

**Price vs. Room Type**

```python
price_room_type = df.groupby('room_type')['price'].mean()
print(price_room_type)
```

Insight: The average price for an entire home/apartment is $ 211.80   while  private rooms  average $89.79, and shared rooms average $70.07. Entire homes are significantly more expensive.

**Reviews per Month vs. Availability**

```python
reviews_vs_availability = df[['reviews_per_month', 'availability_365']].corr()
print(reviews_vs_availability)
```

Insight: There’s a moderate positive correlation (0.163) between the number of reviews per month and the availability of the listing, suggesting that listings with more availability tend to receive more reviews.

**Number of Reviews vs. Price**

```python
reviews_vs_price = df[['number_of_reviews', 'price']].corr()
print(reviews_vs_price)
```

There’s a very weak negative correlation (-0.04) between the number of reviews and price, indicating that lower-priced listings might receive slightly more reviews.

# **Conclusions**
Neighbourhood Popularity: Manhattan and Brooklyn are the most popular areas, with 21,643 and 20,0084 listings respectively, suggesting a high demand for accommodation in these neighborhoods.

Room Type Demand: Entire homes/apartments are the most popular, with an average price of $211.80, showing that guests prefer private, self-contained accommodations.

Price Sensitivity: The median listing price is $106, with a majority of listings priced between $69 and $175. Listings priced within this range are likely to attract more bookings.

Booking Frequency: Listings with higher availability tend to receive more reviews, as shown by the moderate positive correlation (0.163) between availability and reviews per month.

Review Patterns: Lower-priced listings tend to have slightly more reviews, as indicated by the weak negative correlation (-0.04) between the number of reviews and price.

# **Data visualization**

# Distribution of Listings by Neighbourhood Group

Visualized the distribution using a count plot, reordered by descending counts to highlight the most popular neighborhoods.

```python
#Visualize the distribution of listings by neighbourhood_group.
# Calculate the counts for each neighbourhood group
neighbourhood_counts = df['neighbourhood_group'].value_counts()

# Create a new categorical type with the reordered categories
df['neighbourhood_group'] = pd.Categorical(df['neighbourhood_group'],
                                           categories=neighbourhood_counts.index,
                                           ordered=True)

# Plot the count plot with the reordered categories
plt.figure(figsize=(10, 6))
sns.countplot(x='neighbourhood_group', data=df, order=neighbourhood_counts.index)
plt.title('Distribution of Listings by Neighbourhood Group')
plt.show()
```

# **Geographical Distribution of Listings**

# **Correlation Analysis**

Investigated correlations between numerical features to identify potential relationships

```python
#Check correlations between numerical features.
numeric_df = df.select_dtypes(include=[float, int])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
